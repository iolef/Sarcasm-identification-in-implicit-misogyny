# **Humour identification in implicit misogyny**

In this notebook, the classified [AMI 2018 dataset](https://https://amievalita2018.wordpress.com/) is examined to discover if it contains examples of implicit misogyny and/or humour.

# **1. Imports**


```python
import os
import pandas as pd
```

# **2. File upload**

```python
# Selcting the classified dataset that was previously converted into tsv
df_path = os.path.join(os.getcwd(), "classified_dataset.tsv")


# Creating the dataframe
df = pd.read_csv(df_path, delimiter="\t")

df
```

# **3. Dataframes**

A series of dataframes are created, containing the implicit misogyny content, the humorous misogynous content and the humour content in implicit misogyny, so as to be able to save the filtered data if needed.

```python
#creating a dataframe containing only the content labelled as misogynous
misogynous_df = df.loc[df['misogynous'] == 1]
misogynous_df
```

```python
#creating a dataframe from the misogynous dataframe with the implicit_hate content only
implicit_misogyny_df = misogynous_df.loc[misogynous_df['hate'] == 'implicit_hate']
implicit_misogyny_df
```

```python
#creating a dataframe from the misogynous dataframe with the funny content only
humour_in_misogyny_df = misogynous_df.loc[misogynous_df['humour'] == 'funny']
humour_in_misogyny_df
```

```python
#creating a dataframe from the implicit_misogyny dataframe containing also the funny content
humour_in_implicit_misogyny_df = implicit_misogyny_df.loc[implicit_misogyny_df['humour'] == 'funny']
humour_in_implicit_misogyny_df
```

# **4. Percentages**

```python
#counting the number of the samples in each dataframe created
num_samples = len(df)
num_samples_misogynous = len(misogynous_df)
#non abbiamo usato funzione count di pandas ma abbiamo deciso di creare tanti dataframe nuovi per ogni caso
#in modo da poter salvare su disco i dati filtrati
num_samples_implicit_misogyny_df = len(implicit_misogyny_df)
num_samples_humour_in_misogyny_df = len(humour_in_misogyny_df)
num_samples_humour_in_implicit_misogyny_df = len(humour_in_implicit_misogyny_df)


#printing the results
print(f"The total number of instances in the AMI 2018 dataset is {num_samples}")
print(f"The misogynous instances are {num_samples_misogynous}, {100*num_samples_misogynous/num_samples}% of the total")
print(f"The number of instances containing implicit misogyny is {num_samples_implicit_misogyny_df}, {100*num_samples_implicit_misogyny_df/num_samples_misogynous}% of the misogynous content")
print(f"The number of instances containing humour is {num_samples_humour_in_misogyny_df}, {100*num_samples_humour_in_misogyny_df/num_samples_misogynous}% of the misogynous content")
print(f"Humour in implicit misogyny is present in {num_samples_humour_in_implicit_misogyny_df} instances, {100*num_samples_humour_in_implicit_misogyny_df/num_samples_misogynous}% of the misogynous content")
#print(f"sample uno o altro {num_sample_implicit_misogynous_df + num_sample_humour_misogynous_df} {100*(num_sample_implicit_misogynous_df + num_sample_humour_misogynous_df)/num_sample_misogynous}% of the misogynous")
```
